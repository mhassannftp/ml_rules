```python
#importing the nessary liabraries and packages
from PIL import Image
import numpy as np
import matplotlib.pylab as plt
```

```python
#Loading images
im = Image.open("C:/Users/M_Hassan/NIAIS/earthsky.jpg")
print(np.max(im))
```

```python
#set the fig size and show
plt.figure(figsize=(10,10))
plt.imshow(im)
plt.axis('off')
plt.show()
```

```python
#Split the Channels
r, g, b = im.split()
```

```python
#reference Point
r_old = np.linspace(0,255,11)
r_new = [0, 12.75, 25.5,51. ,76.5, 127.5, 178.5, 204., 255, 242.25, 255.]
```

```python
r1 = Image.fromarray((np.reshape(np.interp(np.array(r).ravel(),r_old,r_new),(im.height, im.width))).astype(np.uint8), mode='L')
```

```python
#Show origional shape and with red channel interploation and also display both histogram
plt.figure(figsize=(15,11))
plt.subplot(221)
plt.imshow(im)
plt.title("Original", size=15)
plt.axis("off")

plt.subplot(222)
im1 = Image.merge('RGB', (r1,g,b))
plt.imshow(im1)
plt.axis('off')
plt.title("With Red channel interpolation", size=15)

plt.subplot(223)
plt.hist(np.array(r).ravel())

plt.subplot(224)
plt.hist(np.array(r1).ravel())

plt.show()
```

```python
#Now display red interpolatd image and with transfomation Red, Blue
plt.figure(figsize=(15,11))
plt.subplot(121)
plt.imshow(im1)
plt.title("Red-Interpolated Image", size=15)
plt.axis('off')

b1 = Image.fromarray(np.clip(np.array(b)+ 7.65, 0, 255).astype(np.uint8))
im2 = Image.merge('RGB', (r1,g,b1))

plt.subplot(122)
plt.imshow(im2)
plt.axis('off')
plt.title('With transformation -Red, Blue',size=15)
plt.tight_layout()
plt.show()
```

```python
#Now again Transformation RED, BLUE with orgional image
plt.figure(figsize=(14,11))
plt.subplot(121)
plt.imshow(im)
plt.title("Orignal Image", size=15)
plt.axis('off')

plt.subplot(122)
plt.imshow(im2)
plt.axis('off')
plt.title('With transformation -Red, Blue',size=15)
plt.tight_layout()
plt.show()
```
